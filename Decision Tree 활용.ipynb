```python
import pandas as pd
import matplotlib.pyplot as plt # 그래프 그리는 라이브러리
from sklearn.model_selection import train_test_split # train, test로 나눠주고 셔플
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv("data/mushroom.csv")
```

```python
data.head()
```

### 목표
- 버섯의 특징을 활용해서 독/식용 분류

```python
data.shape #행: 8124 열: 23
```

```python
data.info() # 결측값 확인 할 수 있다.
```

```python
X = data.iloc[:, 1:]
```

```python
y = data.iloc[:, 0]
```

```python
data.describe() # unique: 데이터 중복 제외 했을때 남는 값
```

```python
y.value_counts() # 중복제거된 데이터의 수를 알려줌
```

### 원 핫 인코딩

```python
X_one_hot = pd.get_dummies(X)
```

```python
X_one_hot
```

### 라벨인코딩

```python
X['habitat'].unique() #중복 제거
```

```python
habitat_dic = {'u':0, 'g':1, 'm':2, 'd':3, 'p':4, 'w':5, 'l':6}
```

```python
X['habitat'].map(habitat_dic) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size = 0.3, random_state = 70)
```

```python
tree_model = DecisionTreeClassifier(max_depth=3)
```

```python
tree_model.fit(X_train, y_train)
```

```python
pre = tree_model.predict(X_test)
```

```python
pre
```

```python
metrics.accuracy_score(pre, y_test)
```

```python
tree_model.score(X_test, y_test)
```
